1.	Create a time-series plot showing daily total revenue for three different countries with three countries in one plot 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
dfRevAus = pd.read_csv('tableDailyRevenueAustralia.csv')
dfRevCan = pd.read_csv('tableDailyRevenueCanada.csv')

# Plot a time-series plot showing daily total revenue for three countries 
plt.plot(dfRevAus['date_of_revenue'], dfRevAus['daily_revenue'], '-x', label='Austalia')
plt.plot(dfRevCan['date_of_revenue'], dfRevCan['daily_revenue'], '-o', label='Canada')

plt.title('Daily Revenue of Different Countries over Time')
plt.xlabel('Date')
plt.ylabel('Total Daily Revenue')
plt.xticks(rotation=50)
plt.legend()

plt.show()
```

2.	Carry out a sentiment analysis on movie descriptions adding another column to the data frame that provides a sentiment classification.

```python
def sentiment_detection(sentence):
    positive_words = ('happy', 'sunny', 'positive', 'triumphant', 'optimistic', 'wonderful')
    negative_words = ('sad', 'terrible', 'frightening', 'rainy', 'scary')

    positive = any(sentence.count(i) > 0 for i in positive_words)
    negative = any(sentence.count(i) > 0 for i in negative_words)

    if positive == negative == False:
        return "neutral"
    elif positive != negative:
        return "positive" if positive else "negative"
    else:
        return "mixed"
```

```python
dfFilmDes = pd.read_csv('tableFilmDescription.csv')

dfFilmDes['sentiment'] = dfFilmDes['description'].apply(sentiment_detection)
dfFilmDes.head()
```

3.	Create a visual in the notebook that shows how much revenue is coming from different countries to the DVD store 

```python
# Read the CSV file into a DataFrame
dfRevAllCountries = pd.read_csv('tableTotalRevenueAllCountries.csv')
dfRevAllCountriesSorted = dfRevAllCountries.sort_values('total_revenue',ascending=True)
dfRevAllCountriesSorted.head()

```

```python
# Plot a bar chart showing total revenue by country

plt.figure(figsize=(10, 25))
plt.barh(dfRevAllCountriesSorted['country'], dfRevAllCountriesSorted['total_revenue'])

plt.title('Total Revenue by Country')
plt.ylabel('Country')
plt.xlabel('Total Revenue')
```
